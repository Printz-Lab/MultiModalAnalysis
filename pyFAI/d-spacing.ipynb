# Cubic

```python

import numpy as np
from itertools import product

a = 6.31  # lattice constant in Å (cubic MAPbI3)

# Generate all hkl combinations excluding (0, 0, 0)
hkls = [
    (h, k, l)
    for h, k, l in product(range(3), repeat=3)
    if (h, k, l) != (0, 0, 0)
]

# Compute d-spacing
d_spacing_dict = {}
for hkl in hkls:
    h2k2l2 = hkl[0]**2 + hkl[1]**2 + hkl[2]**2
    if h2k2l2 == 0:
        continue
    d = a / np.sqrt(h2k2l2)
    d = round(d, 4)  # round to 4 decimals for uniqueness
    if d not in d_spacing_dict:
        d_spacing_dict[d] = [hkl]
    else:
        d_spacing_dict[d].append(hkl)

# Sort and display
for d in sorted(d_spacing_dict.keys(), reverse=True):
    hkls = d_spacing_dict[d]
    print(f"hkl:{hkls}, d: {d:.4f}")  # q-value calculation
```

# Tetragonal

```python
def tetragonal_d_spacing(h, k, l, a=8.8, c=12.685):
    """
    Compute the d-spacing (Å) for tetragonal MAPbI3.
    
    Parameters
    ----------
    h, k, l : int
        Miller indices of the plane.
    a : float
        Lattice constant a = b (Å). Default 8.8 Å.
    c : float
        Lattice constant c (Å). Default 12.685 Å.
    
    Returns
    -------
    float
        Interplanar spacing d_hkl in Å.
    """
    return 1.0 / np.sqrt((h**2 + k**2) / a**2 + (l**2) / c**2)
hkls = [
    (h, k, l)
    for h, k, l in product(range(5), repeat=3)
    if (h, k, l) != (0, 0, 0)
]

# Compute d-spacing
tetragonal_d_spacing_dict = {}
for hkl in hkls:
    h2k2l2 = hkl[0]**2 + hkl[1]**2 + hkl[2]**2
    if h2k2l2 == 0:
        continue
    d = tetragonal_d_spacing(hkl[0], hkl[1], hkl[2])
    d = round(d, 4)  # round to 4 decimals for uniqueness
    if d not in tetragonal_d_spacing_dict:
        tetragonal_d_spacing_dict[d] = [hkl]
    else:
        tetragonal_d_spacing_dict[d].append(hkl)

# for d in sorted(tetragonal_d_spacing_dict.keys(), reverse=True):
#     hkls = tetragonal_d_spacing_dict[d]
#     print(f"{d:.4f}")  # Display d-spacing values

for d, x in zip(sorted(tetragonal_d_spacing_dict.keys(), reverse=True), x_values):
    hkls = tetragonal_d_spacing_dict[d]
    print(f"hkl: {hkls}, d: {d:.4f}, q-value: {x/10}")  # q-value calculation
```

# Hexagonal (lead iodide)

```python
import numpy as np

def d_spacing_hexagonal(h, k, l, a=4.56, c=6.98):
    h, k, l = np.array(h), np.array(k), np.array(l)
    term1 = (4/3) * (h**2 + h*k + k**2) / a**2
    term2 = (l**2) / c**2
    d_inv_sq = term1 + term2
    return 1 / np.sqrt(d_inv_sq)

hkls = [
    (h, k, l)
    for h, k, l in product(range(5), repeat=3)
    if (h, k, l) != (0, 0, 0)
]

# Compute d-spacing
hexagonal_d_spacing_dict = {}
for hkl in hkls:
    h2k2l2 = hkl[0]**2 + hkl[1]**2 + hkl[2]**2
    if h2k2l2 == 0:
        continue
    d = d_spacing_hexagonal(hkl[0], hkl[1], hkl[2])
    d = round(d, 4)  # round to 4 decimals for uniqueness
    if d not in hexagonal_d_spacing_dict:
        hexagonal_d_spacing_dict[d] = [hkl]
    else:
        hexagonal_d_spacing_dict[d].append(hkl)

# for d in sorted(tetragonal_d_spacing_dict.keys(), reverse=True):
#     hkls = tetragonal_d_spacing_dict[d]
#     print(f"{d:.4f}")  # Display d-spacing values

for d in sorted(hexagonal_d_spacing_dict.keys(), reverse=True):
    hkls = hexagonal_d_spacing_dict[d]
    print(f"{d:.4f}")  
```
